```python
import pandas as pd
```

```python
hic_genelist = pd.read_csv("hic_sig_genelist.tsv", sep="\t", header=None)
hic_genelist = hic_genelist[0].tolist()
hic_genelist 
```

```python
atac_genelist = pd.read_csv("atac_sig_genelist.tsv", sep="\t", header=None)
atac_genelist = atac_genelist[0].tolist()
atac_genelist
```

```python
rna_genelist = pd.read_csv("rna_sig_genelist.tsv", sep="\t", header=None)
rna_genelist = rna_genelist[0].tolist()
rna_genelist
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Define the sets
atac_genelist_set = set(atac_genelist)
rna_genelist_set = set(rna_genelist)
hic_genelist_set = set(hic_genelist)

# Create the Venn diagram
venn_labels = {'100': len(atac_genelist_set - rna_genelist_set - hic_genelist_set),
               '010': len(rna_genelist_set - atac_genelist_set - hic_genelist_set),
               '001': len(hic_genelist_set - atac_genelist_set - rna_genelist_set),
               '110': len(atac_genelist_set & rna_genelist_set - hic_genelist_set),
               '101': len(atac_genelist_set & hic_genelist_set - rna_genelist_set),
               '011': len(rna_genelist_set & hic_genelist_set - atac_genelist_set),
               '111': len(atac_genelist_set & rna_genelist_set & hic_genelist_set)}

venn = venn3(subsets=venn_labels, set_labels=('ATAC Genes', 'RNA Genes', 'HiC Genes'))

# Display the Venn diagram
plt.savefig("venn_diagram.pdf", bbox_inches='tight')
```
